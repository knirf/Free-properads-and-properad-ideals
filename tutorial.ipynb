Remember that in this case, weight $w$ and arity $n$ are related ($w = n - 1$), thus most of the time we still sort by weight and biarity for nothing, but in a properadic case it is necessary. Let us start by defining the free (pr)operad over a generator, the product, without symmetry. If you want to generate a new properad from zero, delete the file "/Files/Properads/As"

```python
product = Vertex(1, 2, "prod") #This is the product as a Vertex.
generators = product.graph().all_Graph_Perm() #Here I take the graph associated to product, then all possible graphs with permutations associated to this graph.
As = Free_Properad(generators, 'As') #This generates the properad, so far it only has weight 0 and 1 elements.
```

We can print the number of graphs by weight and biarity (which is here decided by the weight).

```python
dic_weight1 = As.sort_by_arity(1)
dic_weight2 = As.sort_by_arity(2)
dic_weight3 = As.sort_by_arity(3)
dic_weight4 = As.sort_by_arity(4)

print("weight 1 :")
for arity in dic_weight1:
    print("arity " + str(arity) + " : " + str(len(dic_weight1[arity])))
    
print("weight 2 :")
for arity in dic_weight2:
    print("arity " + str(arity) + " : " + str(len(dic_weight2[arity])))
    
print("weight 3 :")
for arity in dic_weight3:
    print("arity " + str(arity) + " : " + str(len(dic_weight3[arity])))
    
print("weight 4 :")
for arity in dic_weight4:
    print("arity " + str(arity) + " : " + str(len(dic_weight4[arity])))
```

Now we can print all weight $2$ elements and name them in order to construct the relations.

```python
i = 0
for graph in As.by_weight[2]:
    print(i) #We print the position in the list.
    print(graph)
    i+=1
```

```python
leftcomb123 = As.by_weight[2][0].copy()
leftcomb132 = As.by_weight[2][1].copy()
leftcomb213 = As.by_weight[2][3].copy()
leftcomb231 = As.by_weight[2][2].copy()
leftcomb312 = As.by_weight[2][4].copy()
leftcomb321 = As.by_weight[2][5].copy()

rightcomb123 = As.by_weight[2][8].copy()
rightcomb132 = As.by_weight[2][11].copy()
rightcomb213 = As.by_weight[2][7].copy()
rightcomb231 = As.by_weight[2][10].copy()
rightcomb312 = As.by_weight[2][6].copy()
rightcomb321 = As.by_weight[2][9].copy()
```

We define now the relations which are given by lists of lists of lists of...

```python
associativity123 = [[leftcomb123, 1], [rightcomb123, -1]]
associativity132 = [[leftcomb132, 1], [rightcomb132, -1]]
associativity213 = [[leftcomb213, 1], [rightcomb213, -1]]
associativity231 = [[leftcomb231, 1], [rightcomb231, -1]]
associativity312 = [[leftcomb312, 1], [rightcomb312, -1]]
associativity321 = [[leftcomb321, 1], [rightcomb321, -1]]

associativity = [associativity123, associativity132, associativity213, associativity231, associativity312, associativity321]
```

And the properadic ideal. Here we generate two steps only if there is no step generated, the reason why we cannot do as for free properads is because the relations can be in different weights, not necessarly weight $2$, thus what we do is generate step by step instead of weight by weight.

Thus here if it happens that the ideal has already compute two steps, the next one is really long and not necessarly required, so we don't generate it right now.

```python
As_relations = Ideal_Properad(As, associativity, "As_relations")
if As_relations.generated_steps[-1] <= 0:
    As_relations.generate_next_weight()
```

Now we can again sort by arity, generate matrices and look for dimensions of the relations in weight $2$, $3$ and $4$.

```python
matrices_arity3 = As_relations.generate_matrices(0)
matrices_arity4 = As_relations.generate_matrices(1)

print("Weight 2 : ")
for arity in matrices_arity3:
    print("arity : " + str(arity) + " : " + str(matrices_arity3[arity].rank()))

print("Weight 3 : ")
for arity in matrices_arity4:
    print("arity : " + str(arity) + " : " + str(matrices_arity4[arity].rank()))
```

But this does not give us the representation structure of $As$. In order to get it, one can run representation (or Clifton) matrices. For example in weight $3$ :

```python
weight3 = As_relations.only_one_per_orbite(1)[(1, 4)]
dic_matrices = create_Cliftons(weight3)
```

```python
for couple in dic_matrices:
    print("partitions : " + str(couple) + " : " + str(dic_matrices[couple].rank()))
```

We have each time two partitions, but the first one is a partition of $1$, this is again because we can consider properads, but here we actually only need a partition of $4$.

Here, considering that the free properad over the generator product is $5$ copies of the regular representation of $\mathbb{S}_4$ in weight $3$, we proved that the operad $As$ is regular in weight $3$. To go further in this direction, one look at the following space of relations and ask the question : For which $x \in \mathbb{Q}$ the operad given by these relations is regular ?

```python
Ra.<x> = QQ[]

x_associativity123 = [[leftcomb123, 1], [rightcomb123, -x]]
x_associativity132 = [[leftcomb132, 1], [rightcomb132, -x]]
x_associativity213 = [[leftcomb213, 1], [rightcomb213, -x]]
x_associativity231 = [[leftcomb231, 1], [rightcomb231, -x]]
x_associativity312 = [[leftcomb312, 1], [rightcomb312, -x]]
x_associativity321 = [[leftcomb321, 1], [rightcomb321, -x]]

x_associativity = [x_associativity123, x_associativity132, x_associativity213, x_associativity231, x_associativity312, x_associativity321]
```

In order to answer this question, one can do the same computations as before but with these relations.

```python
x_As_relations = Ideal_Properad(As, x_associativity, "x_As_relations")
if x_As_relations.generated_steps[-1] <= 0:
    x_As_relations.generate_next_weight()
    
matrices_arity3 = As_relations.generate_matrices(0)
matrices_arity4 = As_relations.generate_matrices(1)
    
x_weight3 = x_As_relations.only_one_per_orbite(1)[(1, 4)]
x_dic_matrices = create_Cliftons(x_weight3)
x_dic_matrices_simple = Clifton_matrix_simple(x_dic_matrices)
save_matrices_csv(x_dic_matrices_simple, "tutorial/x_As/", "")

for couple in x_dic_matrices_simple:
    print("partitions " + str(couple) + " : ")
    print(x_dic_matrices_simple[couple][1])
```

This proves, as M. Bremner and V. Dotsenko did but in a more general case, that the only regular properads of this form are for $x = 0$ or $x = 1$.
